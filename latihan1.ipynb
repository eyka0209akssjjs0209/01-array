```python
import numpy as bakso
```

```python

jumlahe=[]
jumlah=int(input("jumlah mahasiswa:"))
jumlahe.append(jumlah)
i=1
nilay=[]
namay=[]
for i in range(1,jumlah):
    nm=input(f"nama mahasiswa ke- {i}:")
    namay.append(nm
                 )
    ni=int(input(f"nilai mahasiswa ke- {i}:"))
    nilay.append(ni)
i=+1
```

```python
kecil=min(nilay)
print("nilai terkecil:", kecil)
besar=max(nilay)
print("nilai terbesar:", besar)
rata2=sum(nilay)/jumlah
print("nilai rata-rata:", rata2)
```

```python
lulus=[score for score in nilay if score >=60] #tidak tahu caranya hitung jadi line satu ini 100% AI, saya tidak paham maksud code nya apa
print("mahasiswa lulus:",len(lulus))
```

```python
grap=bakso.array([[nilay],[namay]])
print(grap)
```

```python
import matplotlib.pyplot as hehe
hehe.bar(range(len(nilay)), nilay)
hehe.title("grafik nilai")
hehe.xlabel("nama")
hehe.ylabel("nilai")
hehe.show()
```

```python

import matplotlib.pyplot as oek_oek
labels = ['lulus', 'gak lulus']
sizes = (len(lulus), (jumlah-len(lulus)))
oek_oek.pie(sizes, labels=labels, autopct='%1.1f%%',
        startangle=90)
oek_oek.title("grafik lulus")
oek_oek.axis('equal')
oek_oek.show()
```
